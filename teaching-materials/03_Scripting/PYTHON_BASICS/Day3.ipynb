```python
#forward index
my_first_object = ('james', 20, 23.78, True, False, 45, 22.1, 'Ice')
print(type(my_first_object))

print(my_first_object[0])
print(my_first_object[1])


#Backward index
print(my_first_object[-1])
print(my_first_object[-3])


# Forward range index
print(my_first_object[:3]) #[start:stop:]
print(my_first_object[::2]) #[start:stop:step:]



#Backward range index
print(my_first_object[::-1])
print(my_first_object[::-2])
print(len(my_first_object))




```

```python
my_tuple = (1,2,3,4,5, 6)

print(my_tuple[1:3])
print(my_tuple[1:5:2])
```
