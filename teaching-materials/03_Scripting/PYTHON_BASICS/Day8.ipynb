```python
# read()
# write()
# with()
# close()
# open()
# seek()
# r --> raw string
# "r" --> denoted to read
# "w" --> denoted to  write and truncate
# "a" --> append
# "w+" --> write and read
# "a+" --> append and read
```

```python
# Reading a file
file = open(r"C:\Users\praise\Downloads\myfile.txt", "r")
content = file.read()
print(content)
file.close()
```

```python
# Writing a file
filepath = r"C:\Users\praise\Downloads\myfile.txt"
file = open(filepath, "w")
file.write("keep on coding")
file.close()
```

```python
# append a file
file = open(filepath, "a")
file.write("\nThe more you code, the better you get at it.")
file.close()
```

```python
# write and read
file = open(filepath, "w+")
file.write("Keep on coding")
file.seek(0)
content = file.read()
print(content)
file.close()
```

```python
# append and read
file = open(filepath, "a+")
file.write("\nPython is easier than all programming language")
file.seek(0)
content = file.read()
print(content)
file.close()
```

```python
# using the with method
with open(filepath, "r") as file:
    content = file.read()
    print(content)
```

```python
with open(filepath, "a+") as file:
    file.write("\nGod is Good")
    file.seek(0)
    content = file.read()
    print(content)
```
