```python
#student average calculator

students = int(input('Enter number of students: '))
total = 0
scores = []

for i in range(students):
    score = float(input("Enter score: "))
    scores.append(score)
    total += score

average = total/students

print('Average score', average)

for s in scores:
    print(s)



```

While loop runs when a condition is true and stops when a condition if false

```python
count = 0
while count <= 10:
    print(count)
    count += 2
```

```python
count = 0
while count >= 10:
    print(count)
    count -= 1
```

```python
# password checker
#create the password variable that holds the correct/fixed password
password = "Nobody@123"
# Creating a variable for attempts a user should have
attempts_left = 3

#' created a user input variable(we are to call it later in the program )
user_input = ''
# While loop condition
while user_input != password and attempts_left > 0:
    user_input = input("Enter your password: ")
    attempts_left -= 1

    if user_input == password:
        print("Access Granted")
    elif attempts_left > 0:
        print(f"Wrong password! You have {attempts_left} attempts left.")
    else:
        print('Access Denied!')
```

```python
import random 

selection = ['Rock', 'Paper', 'Scissors']
user_score = 0
computer_score = 0

while True:
    computer = random.choice(selection)
    user = input('Enter your selection (or exit) ').strip().capitalize()

    if user == 'Exit':
        print(f"Final score - You: {user_score} vs computer: {computer_score}")
        break
    elif user not in selection:
        print('Wrong input! Choose Rock, Paper or Scissors')
    elif user == computer:
        print(f"It's a tie! Computer also chose {computer}")
        user_score += 1
        computer_score +=1
    elif (user == 'Rock' and computer == 'Scissors') or \
         (user == 'Paper' and computer == 'Rock') or \
         (user == 'Scissors' and computer == 'Paper'):
        print(f"You won, computer chose {computer}")
        user_score += 2
    else:
        print(f'You lost, computer chose {computer}')
        computer_score += 2
```
