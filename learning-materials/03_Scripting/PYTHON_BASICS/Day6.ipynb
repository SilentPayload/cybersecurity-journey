```python
# Password strength checker

def check_password(password, min_length):
    if len(password) >= min_length:
        print(password, "- Strong password(Meets the minimum length)")
    else:
        print(password, "- Weak password(Too short!)")

check_password("abc", 8)
check_password("strong", 8)
check_password("Password123", 8)
```

```python
def is_strong_password(password):
    if len(password) >= 8:
        return True
    else:
        return False
    

password1 = "Weak"
password2 = "Securepass@123"

if is_strong_password(password1):
    print(password1, "is a strong password.")
else:
    print(password1, "is a weak password.")


if is_strong_password(password2):
    print(password2, "is a strong password")
else:
    print(password2, "is a weak password")
```

```python
# login system

def check_username(username):
    valid_users = ['admin', 'michael', 'samson']
    if username in valid_users:
        return True
    else:
        return False
    
def check_password_match(password, correct_password):
    if password == correct_password:
        return True
    else:
        return False
    
def login(username, password):
    correct_password = 'somebody'

    if check_username(username) and check_password_match(password, correct_password):
        print("Access Granted - Welcome", username)
    elif not check_username(username):
        print("Access Denied - Username is not found")
    else:
        print("Access Denied - wrong password")    


login("praiz", "SecurePass123")
login("samson", "somebody")
login("michael", "somebody")
login("admin", "somebody")
login("admin", "somebody123")
```
